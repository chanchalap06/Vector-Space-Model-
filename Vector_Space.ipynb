```python
import string
import nltk
import re
import math
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from termcolor import colored
```

```python
stopwords=nltk.corpus.stopwords.words('english')
path="C:/Users/ilaan/Downloads/Corpus-20230203T210935Z-001/Corpus/"
```

```python
import os
my_files=os.listdir(path)
print(my_files)
```

```python
docu_id={}

for i in range(len(my_files)):
    docu_id[my_files[i]]=i
print(docu_id)

inverted_docu_id={}
for i in range(len(my_files)):
    inverted_docu_id[i]=my_files[i]
print(inverted_docu_id)

```

```python
def read_one_file(file_name): #function to read one file
 lines=''
 with open(path+file_name,"r",encoding='utf-8') as f  :
  lines = ' '+lines+f.read()
 return lines
```

```python
def remove_punctuation(line):
  lines="".join([char for char in line if char not in string.punctuation])
  lines=re.sub('[0-9]+','',lines)
  return lines
```

```python
def tokenization_case(text):
  text=text.strip()
  text=re.split('\W+',text)
  text=list([x.lower() for x in text]) #case folding
  #print(text)
  return text
```

```python
def remove_stopwords(text):
  text=[word for word in text if word not in stopwords]
  return text

def stemming(text):
  text_tokens= [nltk.PorterStemmer().stem(word) for word in text]
  return text_tokens

def lemmatizer(text):
  text_tokens= [nltk.WordNetLemmatizer().lemmatize(word) for word in text]
  return text_tokens
```

```python
inverted_index={}

for file in my_files:
 lines=remove_punctuation(read_one_file(file))
 tokens=tokenization_case(lines)
 tokens=lemmatizer(stemming(remove_stopwords(tokens)))
  
 for word in tokens:
    if word not in inverted_index.keys():
        inverted_index[word]=[]
        inverted_index[word].append(1)
        inverted_index[word].append({})
        inverted_index[word][1][docu_id[file]]=1
    else:
       if docu_id[file] not in inverted_index[word][1].keys():
         inverted_index[word][0]+=1
         inverted_index[word][1][docu_id[file]]=1   
       if docu_id[file] in  inverted_index[word][1].keys():
            inverted_index[word][1][docu_id[file]]+=1
       
           
            
print(inverted_index)            
 
```

```python
def query_processing(query):
  query = re.sub(r"\s+"," ",query)
  query_tokens = query.split(' ')
  print('Initial Tokens:',query_tokens)  
  i=0
  unique_tokens=[]  
  while(i<len(query_tokens)):
    #print('yes')

      lines = query_tokens[i]
       
      lines = remove_punctuation(lines)
      tokens = tokenization_case(lines)
     
      tokens = remove_stopwords(tokens)
      tokens = stemming(tokens)
      #print('2',tokens)
      tokens = lemmatizer(tokens)
      if(len(tokens)!=0 and tokens[0]!=''):
       #print('Yes')     
       query_tokens[i]=tokens[0]
       i=i+1
       if tokens[0] not in unique_tokens:
         unique_tokens.append(tokens[0])
    
      else:
        query_tokens.remove(query_tokens[i])
    
            
    
  return query_tokens,unique_tokens
```

```python
N=len(my_files)

def normalize(li):
    #print(li)
    total=0
    for x in li:
        total=total+(x**2)
    mag=math.sqrt(total)
    #print(mag)
    for i in range(len(li)):
        li[i]=li[i]/mag
    #print(li)    
    return li    
    
def query_weight(query_tokens,unique_tokens):
    weights=[]
    for term in unique_tokens :
        #print(term)
        tf=query_tokens.count(term)
        #print('tf_raw=',tf)
        df=inverted_index[term][0]
        #print('data freq=',df)
        if(tf!=0):
         tf=1+math.log10(tf)
         #print('tf=',tf)
        idf=math.log10((N/df))
        #print('idf=',idf)
        weight=tf*idf
        #print('weight=',weight)
        weights.append(weight)
    
    weights=normalize(weights)
    return weights
        
    
        
        
    
```

```python
def document_weight(unique_tokens,document):
    weights=[]
    for term in unique_tokens:
     if document in inverted_index[term][1].keys():
        tf=inverted_index[term][1][document]
     else:
        tf=0
     #print(tf)
     if(tf!=0):
      tf=1+math.log10(tf)
     weights.append(tf)
    #print(weights)
    #print(weights)
    weights=normalize(weights)
    #print(weights)
    return weights
        
            
    
    
    
```

```python
def find_word(lines,word):
  word_index = lines.find(word)
   
  if(word_index==-1):
    #print(first_query_token.capitalize())
    word_index = lines.find(word.capitalize())
  return word,word_index  
```

```python
#"Developing your Zomato business account and profile is a great way to boost your restaurants online reputation"
query=input('Give Query=')
#tokens=query_processing(query)
query_tokens,unique_tokens=query_processing(query)
print('\nTokens after Processing',query_tokens,unique_tokens)
weights_query=query_weight(query_tokens,unique_tokens)
print('\nquery weights=',weights_query)
#print(tokens.count('le'))
score={}
all_doc=set()
doc_term_fre={}
for term in unique_tokens:
    term_doc=inverted_index[term][1].keys()
    #print(term_doc)
    for doc in term_doc:
     if inverted_docu_id[doc] in doc_term_fre.keys():
        doc_term_fre[inverted_docu_id[doc]].append(term)
     else:
        doc_term_fre[inverted_docu_id[doc]]=[term]
    all_doc=all_doc.union(term_doc)
    #
print('\nAll documents with the query terms-->',all_doc)    
for doc in all_doc:
    doc_weight=document_weight(unique_tokens,doc)
    #print(doc,doc_weight)
    s=0
    for i in range(len(weights_query)):
      s=s+weights_query[i]* doc_weight[i]
    score[doc]=s
print('\nDocumnets scores-->',score) 
#print(doc_term_fre)
sorted_value=sorted(doc_term_fre.values(),reverse=True)
sorted_fre = {}

for i in sorted_value:
    for k in  doc_term_fre.keys():
        if  doc_term_fre[k] == i:
            sorted_fre[k] =  doc_term_fre[k]
            
sorted_values = sorted(score.values(),reverse=True) # Sort the values
sorted_score = {}

for i in sorted_values:
    for k in score.keys():
        if score[k] == i:
            sorted_score[k] = score[k]

print('\nAfter sorting -->',sorted_score)
print('\nOutput')
inc=0
for file_num,score in sorted_score.items():
  #print(i)  
  if(inc==10):
        break
  name = inverted_docu_id[file_num]
  print('\033[1;32m')
  print(name,':',score, '\033[0m')
  with open(path+name,"r",encoding='utf-8') as f:
   lines = f.read()
   i=0
   word_index=-1 
   while(word_index==-1 and i<len(query_tokens)):
      word,word_index =find_word(lines,query_tokens[i])
      i=i+1  
   if(word_index>30):    
    substring = lines[word_index-30:word_index+30]
   else:
     substring = lines[word_index:word_index+30]
   space = substring[30:60].find(" ")
   substring = substring.replace("\n"," ")
   if(len(substring)>10):
    print('...',substring[:29],'\033[0;33;1m',' ',word,'\033[0m',substring[30+space:],'...',sep='')
   else:
    print('......',sep='')
  inc=inc+1
#print(sorted_fre)        
```
